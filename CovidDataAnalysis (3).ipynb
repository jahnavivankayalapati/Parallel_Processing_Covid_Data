```python
import numpy as np
import pandas as pd
import time
import threading
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Load your data here (for example, main_data)
main_data = pd.read_csv('Covid Dataset.csv')  # Replace with your dataset
# Encode categorical columns using Label Encoding
label_encoder = LabelEncoder()
for col in main_data.columns:
    main_data[col] = label_encoder.fit_transform(main_data[col])

# Split the data into features and target variable
X = main_data.drop(columns=["COVID-19"])  # Replace "COVID-19" with your target variable
y = main_data["COVID-19"]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Number of threads to use
num_threads = 4  # You can adjust this based on your system capabilities
models = [None] * num_threads
train_times = [0] * num_threads

# Function to train a Random Forest model
def train_model(thread_id, X_split, y_split):
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    train_start_time = time.time()
    model.fit(X_split, y_split)
    train_end_time = time.time()
    models[thread_id] = model
    train_times[thread_id] = train_end_time - train_start_time
    print(f"Thread {thread_id}: Training time: {train_times[thread_id]:.4f} seconds")

# Split the training data among threads
X_train_splits = np.array_split(X_train, num_threads)
y_train_splits = np.array_split(y_train, num_threads)

# Create and start threads
threads = []
for i in range(num_threads):
    thread = threading.Thread(target=train_model, args=(i, X_train_splits[i], y_train_splits[i]))
    threads.append(thread)
    thread.start()

# Wait for all threads to complete
for thread in threads:
    thread.join()

# Evaluate the model on the test set
y_pred = np.zeros(y_test.shape)
for model in models:
    y_pred += model.predict(X_test)

y_pred = np.where(y_pred > (num_threads / 2), 1, 0)  # Majority voting

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Overall Accuracy:", accuracy)

# Get user input
user_input_df = pd.DataFrame({
    'Breathing Problem': [0],  # Example input (1: Yes, 0: No)
    'Fever': [0],
    'Dry Cough': [1],
    'Sore throat': [0],
    'Running Nose': [0],
    'Asthma': [0],
    'Chronic Lung Disease': [0],
    'Headache': [0],
    'Heart Disease': [0],
    'Diabetes': [0],
    'Hyper Tension': [0],
    'Fatigue': [1],
    'Gastrointestinal': [0],
    'Abroad travel': [0],
    'Contact with COVID Patient': [0],
    'Attended Large Gathering': [0],
    'Visited Public Exposed Places': [0],
    'Family working in Public Exposed Places': [0],
    'Wearing Masks': [1],
    'Sanitization from Market': [1]
})

# Ensure user_input_df contains all required columns in the correct order
required_columns = X_train.columns
user_input_df = user_input_df.reindex(columns=required_columns, fill_value=0)

# Start timing for testing
test_start_time = time.time()
user_pred = np.zeros(user_input_df.shape[0])
for model in models:
    user_pred += model.predict(user_input_df)
user_pred = np.where(user_pred > (num_threads / 2), 1, 0)  # Majority voting
test_end_time = time.time()

# Calculate and print testing time
test_time = test_end_time - test_start_time
print(f"Testing time: {test_time:.4f} seconds")

# Output results as Yes or No
prediction = 'Yes' if user_pred[0] == 1 else 'No'
print("Predicted Disease Present:", prediction)
```
